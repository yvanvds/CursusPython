## 1. Van een continue wereld naar meetpunten

Wanneer we in de fysica spreken over een trilling, bijvoorbeeld een geluidsgolf, dan beschrijven we een grootheid die continu verandert in de tijd. Denk aan luchtdruk: op elk moment bestaat er een waarde, en die verandert vloeiend.

Een computer kan echter niet werken met een oneindig aantal tijdstippen. Daarom moeten we de werkelijkheid discreet maken: we kiezen een beperkt aantal momenten waarop we meten.

Zo’n meetpunt noemen we een **sample**.

Als je bijvoorbeeld tien keer per seconde meet, dan krijg je tien samples per seconde. Dit noemen we de **samplerate**.

### Tijd opdelen in meetpunten

````python
import numpy as np

SAMPLERATE = 10
samples = np.linspace(0, 1, SAMPLERATE, endpoint=False)

print(samples)
````

Hier gebeurt het volgende:

* Je kiest een tijdsinterval van 0 tot 1 seconde
* Je verdeelt dat interval in 10 gelijke stukken
* Elk getal in `samples` stelt een tijdstip voor waarop je een meting doet

Het is belangrijk om te begrijpen dat deze lijst niet de golf zelf is, maar enkel de momenten waarop je kijkt naar die golf.

#### Waarom `endpoint=False`?

Je zou kunnen denken dat het logisch is om zowel 0 als 1 op te nemen. Toch doen we dat niet.

De reden is dat veel signalen **periodiek** zijn. Dat betekent dat het punt op tijdstip 1 seconde identiek is aan dat op 0 seconden (het begin van de volgende periode). Als je beide opneemt, meet je dus eigenlijk hetzelfde punt twee keer.

## 2. Van tijd naar een gemeten waarde

Tot nu toe heb je enkel tijdstippen. Maar een meting bestaat uit twee dingen:

* wanneer je meet (tijd)
* wat je meet (waarde)

Samen vormen die een **signaal**.

In de wiskunde beschrijven we dat als een functie:
de waarde hangt af van de tijd.


### Voorbeeld: een lineaire stijging

Stel dat de druk gedurende één seconde gelijkmatig stijgt van 0 naar 10.


````python
import numpy as np

SAMPLERATE = 10
samples = np.linspace(0, 1, SAMPLERATE, endpoint=False)
values = samples * 10

print(samples)
print(values)
````

Wat hier gebeurt, is eenvoudig maar belangrijk:

* `samples` loopt van 0 tot 1
* door te vermenigvuldigen met 10 krijg je waarden van 0 tot 9

### Visualisatie

Visualisatie helpt hier enorm om de impact te zien. We gebruiken opnieuw matplotlib.


````python
import numpy as np
import matplotlib.pyplot as plt

SAMPLERATE = 10
samples = np.linspace(0, 1, SAMPLERATE, endpoint=False)
values = samples * 10

plt.plot(samples, values, marker="o")
plt.ylim(0, 10)
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Sample Values Over Time")
plt.show()
````

De grafiek toont afzonderlijke punten. De lijn die je ziet, is een visuele hulp. In werkelijkheid “kent” de computer enkel de punten.

## 3. Periodieke signalen en de sinus

Veel fysische trillingen zijn **periodiek**. Dat betekent dat ze zichzelf blijven herhalen. De eenvoudigste wiskundige beschrijving van zo’n trilling is de sinusfunctie.

## Eerste poging


````python
import numpy as np
import matplotlib.pyplot as plt

SAMPLERATE = 10
samples = np.linspace(0, 1, SAMPLERATE, endpoint=False)
values = np.sin(samples)

plt.plot(samples, values, marker="o")
plt.ylim(-1, 1)
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Sample Values Over Time")
plt.show()
````

Dit levert echter geen volledige golf op. Dat komt omdat de sinusfunctie werkt met **radialen**, niet met seconden.
Een volledige sinusgolf komt overeen met een hoek van:

$
0 \rightarrow 2\pi
$

Als je dus één volledige golf wil in één seconde, moet je de tijd herschalen.

### Correcte voorstelling


````python
import numpy as np
import matplotlib.pyplot as plt

SAMPLERATE = 10
samples = np.linspace(0, 1, SAMPLERATE, endpoint=False)
values = np.sin(2 * np.pi * samples)

plt.plot(samples, values, marker="o")
plt.ylim(-1, 1)
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Sample Values Over Time")
plt.show()
````

Nu gebeurt het volgende:

* `samples` loopt van 0 tot 1 seconde
* vermenigvuldigen met (2\pi) zorgt ervoor dat de sinus één volledige cyclus doorloopt

### Frequentie

De frequentie geeft aan hoeveel golven je hebt per seconde.

De algemene formule wordt nu:

$
y = \sin(2\pi f t)
$

Hierin is:

* (f) de frequentie (in Hertz)
* (t) de tijd


* Bij (f = 1) zie je één golf per seconde
* Bij (f = 2) zie je twee golven per seconde



````python
import numpy as np
import matplotlib.pyplot as plt

SAMPLERATE = 10
samples = np.linspace(0, 1, SAMPLERATE, endpoint=False)
values = np.sin(2 * np.pi * samples)

plt.plot(samples, values, marker="o")
plt.ylim(-1, 1)
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Sample Values Over Time")
plt.show()
````

## 4. Samplerate en frequentie

Wanneer je een golf probeert weer te geven, moet je voldoende meetpunten hebben. Als je te weinig meet, krijg je een vervormd beeld. Dit verschijnsel heet **aliasing**.
Om een golf correct te meten, heb je minstens twee meetpunten per periode nodig. In de praktijk is dat niet voldoende voor een mooie grafiek; je hebt er veel meer nodig.

---

## Experiment
We experimenteren nu met hogere samplerates. Bekijk even hoe de grafiek er nu uitziet.


````python
import numpy as np 
import matplotlib.pyplot as plt

SAMPLERATE = 1000
FREQUENCY = 42
samples = np.linspace(0, 1, SAMPLERATE, endpoint=False)
values = np.sin(2 * np.pi * samples * FREQUENCY)

plt.plot(samples, values, marker="o")
# plt.xlim(0, 0.1) # x-as van 0 tot 0.1 seconden
plt.ylim(-1, 1) # y-as van -1 tot 1
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Sample Values Over Time")
plt.show()
````

We hebben nu wel een hogere samplerate, maar we kunnen die niet goed wergeven. Maar de golf die we tonen is periodiek. Een klein deel van de golf tonen is dus al genoeg. Dat kan je doen met `plt.xlim(0, 0.1)`

```python
SAMPLERATE = 1000
FREQUENCY = 42
```

Met deze waarden:

* neem je 1000 samples per seconde
* heb je 42 golven per seconde

Dat betekent dat je ongeveer 24 meetpunten per golf hebt, wat voldoende is voor een duidelijke voorstelling.

## 5. Amplitude

De amplitude bepaalt hoe sterk de trilling is. In een geluidsgolf komt dit overeen met het volume.

### Implementatie

````python
import numpy as np 
import matplotlib.pyplot as plt

SAMPLERATE = 1000
FREQUENCY = 42
AMPLITUDE = 0.5
samples = np.linspace(0, 1, SAMPLERATE, endpoint=False)
values = np.sin(2 * np.pi * samples * FREQUENCY) * AMPLITUDE

plt.plot(samples, values)
plt.xlim(0, 0.1) # x-as van 0 tot 0.1 seconden
plt.ylim(-1, 1) # y-as van -1 tot 1
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Sample Values Over Time")
plt.show()
````



Met amplitude verandert enkel de hoogte van de golf, niet de vorm.

## 6. Demping

In de realiteit verliest een trilling energie. Een aangeslagen snaar blijft niet eeuwig trillen. En wanneer je een steen in het water gooit, dan blijven de golven niet oneindig verder gaan.

## Lineaire demping

We zouden dit kunnen uitdrukken met een eenvoudige lineaire curve.



````python
import numpy as np 
import matplotlib.pyplot as plt

SAMPLERATE = 1000
FREQUENCY = 42
AMPLITUDE = 0.5
DAMPENING_FACTOR = 0.5
samples = np.linspace(0, 1, SAMPLERATE, endpoint=False)
values = np.sin(2 * np.pi * samples * FREQUENCY) * AMPLITUDE * (1 - samples * DAMPENING_FACTOR)

plt.plot(samples, values)
plt.xlim(0, 1) # x-as van 0 tot 0.1 seconden
plt.ylim(-1, 1) # y-as van -1 tot 1
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Sample Values Over Time")
plt.show()
````

* Pas de limiet op de X as aan zodat je terug 1 seconde ziet.
* Wijzig dampening factor naar 2. Wat gaat er mis?
* Ken je een wiskundige functie die hier beter zou werken?

## Exponentiële demping

Een realistischer model is:

$
A(t) = e^{-kt}
$

````python
import numpy as np 
import matplotlib.pyplot as plt

SAMPLERATE = 1000
FREQUENCY = 42
AMPLITUDE = 0.5
DAMPENING_FACTOR = 5
samples = np.linspace(0, 1, SAMPLERATE, endpoint=False)
values = np.sin(2 * np.pi * samples * FREQUENCY) * AMPLITUDE * np.exp(-samples * DAMPENING_FACTOR)

plt.plot(samples, values)
plt.xlim(0, 1) # x-as van 0 tot 0.1 seconden
plt.ylim(-1, 1) # y-as van -1 tot 1
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Sample Values Over Time")
plt.show()
````

Deze functie:

* blijft altijd positief
* daalt snel in het begin en daarna trager
* komt overeen met veel echte fysische systemen

## 7. Superpositie van golven

In de praktijk bestaat een signaal zelden uit één enkele frequentie. Een geluid, maar ook andere signalen die je bestudeert, bestaat meestal uit een combinatie van meerdere trillingen.

We kunnen makkelijk twee golven tegelijk tonen met matplotlib. Maar komt dit overeen met de realiteit?



````python
import numpy as np 
import matplotlib.pyplot as plt

SAMPLERATE = 1000
FREQUENCY1 = 42
FREQUENCY2 = 21
AMPLITUDE = 0.3
samples = np.linspace(0, 1, SAMPLERATE, endpoint=False)
wave1 = np.sin(2 * np.pi * samples * FREQUENCY1) * AMPLITUDE
wave2 = np.sin(2 * np.pi * samples * FREQUENCY2) * AMPLITUDE

plt.plot(samples, wave1, label="Wave 1 (42 Hz)", alpha=0.5)
plt.plot(samples, wave2, label="Wave 2 (21 Hz)", alpha=0.5)
plt.xlim(0, 0.2) # x-as van 0 tot 0.1 seconden
plt.ylim(-1, 1) # y-as van -1 tot 1
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Sample Values Over Time")
plt.show()
````


Je hebt niet meerdere drukken tegelijk, maar één druk die ontstaat door het optellen van verschillende golven.

## 8. Van grafiek naar geluid

Tot nu toe heb je een visuele voorstelling van een golf. Maar een geluidsgolf is eigenlijk precies hetzelfde type signaal. Om live met geluid te werken is Python niet snel genoeg. Maar een kort geluid genereren is zeker mogelijk. 

**Als je sounddevice nog niet geinstalleerd hebt kan je dat doen met `pip install sounddevice`.**

### Geluid genereren
(De code werkt niet in je browser, want die kent de module sounddevice niet.)


````python
import sounddevice as sd

SAMPLERATE = 1000
FREQUENCY = 400

samples = np.linspace(0, 1, SAMPLERATE, endpoint=False)
wave = np.sin(2 * np.pi * samples * FREQUENCY)

sd.play(wave, SAMPLERATE)
sd.wait()
````

De computer stuurt de waarden van de golf naar je luidspreker. Die zet ze om in drukverschillen in de lucht.
