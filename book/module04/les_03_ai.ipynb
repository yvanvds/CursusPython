# AI als code-assistent

In deze les onderzoeken we hoe je AI kan gebruiken om Python-code te schrijven voor data-analyse. AI kan veel werk versnellen, maar het blijft belangrijk dat je **begrijpt wat de code doet en controleert of de resultaten correct zijn**.

We starten met een voorbeeld van **vibe coding**, en gaan daarna over naar een meer gecontroleerde manier van AI-gebruik bij Python-analyse.

---

## 1. Vibe coding: een volledige app met één prompt

AI kan tegenwoordig volledige applicaties genereren op basis van een korte beschrijving. Dat wordt vaak **vibe coding** genoemd.

Bij vibe coding beschrijf je een idee en laat je AI de software bouwen. Voorbeeld van een prompt:


````python
With this dataset I need an app which shows today's earthquakes on a global map.
The interface should show bigger dots for bigger earthquakes and let me filter them on size and time.

Use this dataset as a source:

https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.geojson
````

Dit is een dataset met aardbevingen van de laatste 24 uur. De gegenereerde applicatie toont:

* een wereldkaart
* een punt voor elke aardbeving
* grotere punten voor grotere magnitude
* filters op grootte en tijd

De applicatie werkt en ziet er professioneel uit: [https://seismic-pulse-map.base44.app/]


### Waarom dit indrukwekkend is

Met één prompt kan AI:

* een interface bouwen
* een kaart tonen
* data ophalen uit een API
* interactieve filters toevoegen

Dit zou vroeger **uren of dagen programmeerwerk** hebben gekost. AI kan dus zeer snel **bruikbare software genereren**.


### Maar er is ook een probleem

Wanneer we de code bekijken, zien we iets anders: [https://app.base44.com/apps/69adb0386e639aca802b97a5/editor/workspace/code?filePath=src%2Fpages%2FEarthquakes.jsx] (Als je deze code zelf wil bekijken zal je een (gratis) account moeten maken en voor 'clone this app' moeten kiezen.)

De applicatie bestaat uit:

* tientallen bestanden
* verschillende componenten
* veel JavaScript of TypeScript
* externe libraries

De codebasis is dus **groot en complex**. Voor iemand die geen ervaren softwareontwikkelaar is, is het moeilijk om:

* de volledige code te begrijpen
* fouten te vinden
* de betrouwbaarheid te controleren

### Conclusie

Vibe coding is zeer nuttig om:

* prototypes te maken
* visualisaties te bouwen
* snel ideeën uit te testen

Maar wanneer je **betrouwbare analyse** nodig hebt, is deze aanpak riskant.
Je hebt dan weinig controle over wat de code precies doet.bIn wetenschappelijk werk is controle belangrijker dan snelheid.

## 4. De structuur van een goede AI-workflow

Wanneer je AI gebruikt voor programmeren, is het belangrijk om systematisch te werken. Een goede workflow bestaat uit meerdere stappen.

### Stap 1 – Begrijp het probleem

Voordat je AI gebruikt, moet je zelf weten wat je wil onderzoeken.

Bijvoorbeeld:

* hoeveel aardbevingen zijn er vandaag?
* wat is de grootste magnitude?
* waar gebeuren de meeste aardbevingen?

Zonder duidelijke vraag kan AI ook geen goede analyse maken.

### Stap 2 – Inspecteer de data

Bekijk eerst hoe de data eruit ziet.

Bijvoorbeeld:

* welke kolommen bestaan er?
* welke types data zitten erin?
* hoeveel rijen zijn er?

Bij JSON-data moet je vaak ook de **structuur van de gegevens** bekijken.

### Stap 3 – Gebruik AI om code te genereren

Nu kan je AI vragen om een script te schrijven. Daarin ben je best zo specifiek mogelijk. AI helpt hier bij de **implementatie van de code**. Deze stap bespreken we dadelijk nog in meer detail.

### Stap 4 – Controleer de resultaten

Dit is een cruciale stap. AI kan code schrijven die werkt, maar toch **verkeerde resultaten geeft**. Daarom moet je resultaten altijd controleren. Een eenvoudige eerste controle is om data in de terminal te tonen.

Bijvoorbeeld:

```python
print(df.head())
```

Of:

```python
print(data)
```

Door data te tonen in de terminal kan je:

* controleren of de data correct werd ingelezen
* zien of de structuur klopt
* controleren of waarden logisch lijken

Dit is vaak de **eerste stap om fouten te ontdekken**.

Andere controles kunnen bijvoorbeeld zijn:

* zelf een kleine berekening uitvoeren
* extreme waarden controleren (minimum en maximum)
* geef dezelfde opdracht aan een andere AI

## 5. Voorbereidende vragen

Voordat je AI vraagt om code te schrijven, is het vaak nuttig om eerst **enkele voorbereidende vragen** te stellen. Daarmee verzamel je informatie die helpt om betere code te krijgen.

AI kan namelijk vaak meerdere oplossingen voorstellen. Door eerst te vragen **welke aanpak het meest geschikt is**, krijg je meer inzicht in de beschikbare tools.

### Vraag naar geschikte libraries

Een eerste nuttige vraag is bijvoorbeeld:

```
Welke Python libraries zijn geschikt om data over aardbevingen te analyseren 
en weer te geven in een grafiek?
```

AI kan bijvoorbeeld antwoorden met:

* **pandas** → data verwerken en analyseren
* **matplotlib** → grafieken maken
* **seaborn** → statistische visualisaties
* **plotly** → interactieve grafieken
* **cartopy** → betere wereldkaarten

Zo krijg je een overzicht van mogelijke tools.

### Vraag naar voor- en nadelen

Daarna kan je vragen:

```
Wat zijn de voordelen en nadelen van cartopy in dit geval?
```

Dit helpt om te begrijpen:

* waarom bepaalde libraries bestaan
* wanneer ze nuttig zijn
* wanneer ze minder geschikt zijn

### Vraag naar een mogelijke aanpak

Een andere nuttige voorbereidende vraag is:

```
Hoe zou je dit aanpakken als ik 'x' wil analyseren?
```

AI kan dan eerst uitleggen:

1. hoe de data wordt ingelezen
2. hoe de structuur bekeken kan worden
3. welke analyses mogelijk zijn
4. welke grafieken nuttig zijn

Zo krijg je een **stappenplan** voordat de code wordt geschreven.

### Waarom voorbereidende vragen nuttig zijn

Deze aanpak heeft verschillende voordelen:

* je begrijpt beter wat AI doet
* je krijgt overzicht over mogelijke oplossingen
* je vermijdt dat AI meteen complexe code genereert
* De AI kijkt ook naar de dialoog voor je code. Door vragen te stellen krijg je meer controle.

## 6. Goede en slechte prompts

De kwaliteit van AI-code hangt sterk af van de **prompt** die je gebruikt. Meestal bevat een prompt te weinig informatie. Vermeld altijd exact wat het doel is en op welke manier je dit wil bereiken. 

### Voorbeeld van een slechte prompt


````python
https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.geojson

Schrijf python code om elke aardbeving op een wereldkaart te tonen.
````


Dit is een slechte prompt omdat ze **te weinig informatie** bevat.

AI weet bijvoorbeeld niet:

* welke kolommen er bestaan
* hoe je deze data wil tonen
* welke libraries gebruikt moeten worden

## Voorbeeld van een betere prompt


````python
https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.geojson

Via deze url kan je alle aardbevingen van vandaag opvragen. 
Schrijf python code om elke aardbeving op een wereldkaart te tonen. 
Bekijk deze dataset om te zien welke data je nodig hebt. Ik wil een punt op elke 
locatie en de grootte van dat punt moet afhankelijk zijn van de magnitude ('mag').
````


Deze prompt is beter omdat ze duidelijk aangeeft:

* dat de AI de dataset moet analyseren
* welke data je nodig hebt
* wat er precies getoond moet worden

## Een nog betere prompt

Een goede prompt bevat vaak meerdere elementen:

* beschrijving van de dataset
* doel van de analyse
* gewenste libraries
* gewenste output

Bijvoorbeeld:


````python
https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.geojson

Via deze url kan je alle aardbevingen van vandaag opvragen. 
Schrijf python code om elke aardbeving op een wereldkaart te tonen. 
Bekijk deze dataset om te zien welke data je nodig hebt. 
Gebruik requests om de data in te lezen. Daarna pandas en Matplotlib om een grafiek te tonen. 
Toon de een wereldkaart op de achtergrond via cartopy, maar zonder de landen zichtbaar te maken. 
Enkel de kustlijn is belangrijk. 

Ik wil een punt op elke locatie en de grootte van dat punt moet afhankelijk zijn van de 
magnitude ('mag'). Bij een magnitude kleiner dan 2 moet het punt groen zijn. 
Kleiner dan 4 betekent een oranje punt en alles daarboven wordt een rood punt.
````

Dit leidt meestal tot duidelijkere en bruikbare code.

## 7. Comments en aanpasbare variabelen

Wanneer AI code schrijft, is het nuttig om te vragen dat de code **duidelijk gedocumenteerd** wordt.

Je kan bijvoorbeeld expliciet vragen:

```
Voeg commentaar toe bij elk deel van de code die uitlegt wat die code doet.
```

Commentaar helpt om:

* de code sneller te begrijpen
* fouten gemakkelijker te vinden
* later aanpassingen te maken

### Variabelen die gemakkelijk aangepast kunnen worden

Sommige waarden in een script wil je later snel kunnen veranderen.

Bijvoorbeeld:

* een minimum magnitude
* een tijdsinterval
* een data-bron

Het is handig om zulke waarden bovenaan in het script te plaatsen.

Bijvoorbeeld:

```python
MIN_MAGNITUDE = 4
```

De rest van de code gebruikt dan deze variabelen. Dit heeft een groot voordeel: je kan de output van het programma snel veranderen. Als je weet dat je dit nodig hebt, omdat je bijvoorbeeld naar specifieke informatie wil kunnen zoeken, dan is het een goed idee om AI dadelijk te vragen om die waarden bovenaan je script te zetten.

Je kan AI bijvoorbeeld vragen:

```
Ik wil in het script ook eenvoudig de minimum getoonde magnitude kunnen aanpassen, 
en ook de data kunnen beperken tot een specifiek uur. Zorg dat die waarden boven 
in het script aangepast kunnen worden.
```

## 8. Typische fouten bij AI-gebruik

Hoewel AI zeer nuttig kan zijn, maakt het regelmatig fouten. Daarom is het belangrijk om kritisch te blijven. Hier zijn enkele veel voorkomende problemen.

### Datasetstructuur niet vermelden

AI kan de structuur van een dataset niet raden. Wanneer je bijvoorbeeld schrijft:

```
Analyse this dataset.
```

zal AI vaak **kolomnamen verzinnen**. Als die kolommen niet bestaan, krijg je een foutmelding zoals:

```
KeyError
```

Daarom is het beter om de structuur te tonen of de kolomnamen te vermelden in de prompt.


### Libraries die niet geïnstalleerd zijn

AI gebruikt soms libraries die niet op jouw computer staan. Bijvoorbeeld:

```
ModuleNotFoundError
```


### Verkeerde interpretatie van data

AI begrijpt de betekenis van data niet echt.

Bijvoorbeeld:

* een datum interpreteren als tekst
* een numerieke waarde als categorie behandelen
* verkeerde grafieken maken

Daarom moet je altijd controleren of:

* de data correct werd gelezen
* de analyse logisch is
* de grafiek overeenkomt met de data.




## 9. Debuggen met AI

Wanneer je code uitvoert, krijg je soms een foutmelding. Dat is normaal bij programmeren. Fouten maken hoort bij het proces. AI kan in zulke situaties een nuttige **debug-assistent** zijn.

Debuggen betekent: **de oorzaak van een fout zoeken en oplossen**.

Wanneer Python een foutmelding geeft, kan je die rechtstreeks aan AI tonen. Voorbeeld van een foutmelding:

```
KeyError: 'mag'
```

Dat betekent niet dat je helemaal opnieuw moet beginnen. Plak je foutmelding in de AI chat en vraag naar de oorzaak:

```
Deze code geeft me de volgende fout:

KeyError: 'mag'

Wat gaat er mis?
```
