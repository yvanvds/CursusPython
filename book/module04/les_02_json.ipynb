# Werken met JSON

## Stap 1 – JSON begrijpen als datastructuur

### 1. Wat is JSON?

JSON staat voor **JavaScript Object Notation**.
Het is een tekstformaat om gegevens gestructureerd op te slaan of te versturen via het internet.

Belangrijk:
JSON is **geen Python-code**.
Het is gewoon tekst met een vaste structuur.

JSON wordt gebruikt voor:

* Web-API’s
* Configuratiebestanden
* Datacommunicatie tussen servers en programma’s


### 2. Eenvoudige JSON

Bekijk dit voorbeeld:

```json
{
  "station": "Brussel-Centraal",
  "temperature": 12.7,
  "humidity": 81,
  "timestamp": 1709385600,
  "status": "online"
}
```

#### Wat zie je?

Een JSON-object wordt omgeven door `{ }`.

Binnenin staan **key/value-paren**:

```
"key": value
```

Elke key:

* staat tussen dubbele aanhalingstekens
* is een string

Elke value kan verschillende types hebben.

#### Datatypes herkennen

In dit voorbeeld komen meerdere types voor:

| Key         | Type           | Voorbeeldwaarde    |
| ----------- | -------------- | ------------------ |
| station     | string         | "Brussel-Centraal" |
| temperature | number (float) | 12.7               |
| humidity    | number (int)   | 81                 |
| timestamp   | number (int)   | 1709385600         |
| status      | string         | "online"           |

Belangrijk inzicht:

JSON ondersteunt:

* string
* number (int of float)
* boolean (true/false)
* null
* object
* array

#### Waarom typecontrole belangrijk is

In data-analyse is het essentieel om te weten welk type je voor je hebt.

Voorbeeld:

* Een temperatuur van `12.7` kan je gebruiken in berekeningen.
* Een temperatuur van `"12.7"` (met aanhalingstekens) is een string.
* Met een string kan je niet rekenen zonder eerst om te zetten.

Later in deze les zal je zien dat sommige datasets getallen als string opslaan.
Dan moet je die expliciet omzetten naar een numeriek type.

Altijd eerst kijken naar:

* De structuur
* Het type van de waarden

Pas daarna begin je te programmeren.

#### Wat is een timestamp?

In veel JSON-bestanden kom je een veld tegen met de naam `"timestamp"`.

Een **timestamp** is een geheel getal (integer), maar het is geen “gewoon” getal.
Het stelt een specifiek moment in de tijd voor.

##### Unix-timestamp

De meest gebruikte vorm is de **Unix-timestamp**. Dit is:

> Het aantal seconden dat verstreken is sinds
> 1 januari 1970 om 00:00:00 (UTC).

Voorbeeld:

```json
{
  "timestamp": 1709385600
}
```

Dit getal betekent:

* Tel 1 709 385 600 seconden op
* Vanaf 1 januari 1970 (UTC)
* Je krijgt één exact tijdstip

##### Waarom gebruikt men dit systeem?

Omdat:

* Een integer eenvoudig te verwerken is
* Het onafhankelijk is van taal en regio
* Er geen ambiguïteit is (geen probleem met datumnotatie zoals 03/04/2024)

Een timestamp is dus:

* Objectief
* Universeel
* Goed geschikt voor berekeningen

##### Wat kan je ermee doen?

Met timestamps kan je:

* Tijdverschillen berekenen
* Snelheden bepalen
* Data chronologisch sorteren

Bijvoorbeeld:

```text
verschil = timestamp2 - timestamp1
```

Dat verschil is het aantal seconden tussen twee metingen.


### 3. Geneste structuur

Nu bekijken we een complexer voorbeeld:

```json
{
  "city": "Brussel",
  "coordinates": {
    "latitude": 50.8503,
    "longitude": 4.3517
  },
  "measurements": [
    {
      "time": 1709385600,
      "temperature": 12.7
    },
    {
      "time": 1709389200,
      "temperature": 13.1
    }
  ]
}
```

#### Wat is hier nieuw?

We zien twee nieuwe elementen:

##### Genest object

```
"coordinates": {
    "latitude": 50.8503,
    "longitude": 4.3517
}
```

Dit is een object binnen een object.

##### Array (lijst)

```
"measurements": [
    {...},
    {...}
]
```

Een array wordt omgeven door `[ ]`.

Hier bevat de array meerdere objecten.

### 4. Structuur analyseren

Als je een JSON-bestand bekijkt, stel je jezelf altijd eerst deze vragen:

1. Is dit een object of een array?
2. Welke keys bestaan er?
3. Welke types hebben de values?
4. Zijn er geneste structuren?

Pas daarna schrijf je code. Bijvoorbeeld:

* Wil je alle temperaturen?
  → Dan moet je door de array `"measurements"` lopen.

* Wil je de latitude?
  → Dan moet je eerst naar `"coordinates"` gaan.

Data-analyse begint niet met programmeren,
maar met **kijken en begrijpen van de datastructuur**.

## Stap 3 – JSON ophalen in Python

In stap 2 zag je dat JSON vaak online beschikbaar is via een URL.

We gebruiken daarvoor de library **requests**.

### 1 Een eerste API in Python: wie is er in de ruimte?

We starten met deze API:

```json
http://api.open-notify.org/astros.json
```

Die geeft informatie terug over het aantal mensen in de ruimte en wie dat zijn.

#### Startscript

```python
import requests

url = "http://api.open-notify.org/astros.json"
response = requests.get(url)

data = response.json()
number = data['number']

print("Er zijn momenteel " + str(number) + " mensen in de ruimte.")

for person in data['people']:
    # toon nu de naam van elke persoon
    print("??")
```

---

#### Wat gebeurt er?

##### 1) `import requests`

`requests` is een Python-bibliotheek om HTTP-aanvragen te doen naar een webserver.

Je gebruikt dit wanneer je data wil ophalen van een URL.

##### 2) URL instellen

```python
url = "http://api.open-notify.org/astros.json"
```

We bewaren het webadres als string in een variabele.
Dat is handig, want dan kan je later snel een andere API gebruiken.

##### 3) Data ophalen met `get`

```python
response = requests.get(url)
```

Dit stuurt een HTTP GET-request naar de server.

Het resultaat komt in `response`.
Dat is een object met o.a.:

* statuscode (ok of fout)
* de tekst die de server terugstuurt (hier: JSON)

##### 4) JSON omzetten naar Python-structuur

```python
data = response.json()
```

De server stuurt JSON-tekst terug.
Met `.json()` zet `requests` die tekst om naar Python-structuren:

* JSON object → Python dictionary
* JSON array → Python list

Vanaf nu kan je dus met `data[...]` werken zoals bij een dictionary.

##### 5) Een waarde uit de dictionary halen

```python
number = data['number']
```

In de JSON staat een key `"number"`.
Die value is een integer.

##### 6) Printen met string + integer

```python
print("Er zijn momenteel " + str(number) + " mensen in de ruimte.")
```

`print()` kan niet zomaar een string met een integer “optellen”.
Daarom zetten we de integer om naar string met `str(number)`.

##### 7) Door een lijst lopen

```python
for person in data['people']:
```

`data['people']` is een **lijst** (array in JSON).
Elk element van die lijst is een object/dictionary met info over één persoon. 
In elke iteratie is `person` dus een dictionary.

#### 5 Deze data verandert (live data)

Het ISS beweegt voortdurend.
De API geeft dus niet altijd dezelfde waarden terug.

Je kan je code zo aanpassen dat je elke 5 seconden opnieuw een meting doet.

```python
import requests
import time

url = "http://api.open-notify.org/iss-now.json"

while True:
    response = requests.get(url)
    data = response.json()

    lat = data["iss_position"]["latitude"]
    lon = data["iss_position"]["longitude"]

    print("Latitude:", lat, "| Longitude:", lon)

    time.sleep(5)
```

Wat gebeurt hier?

* `while True:` zorgt voor een eindeloze herhaling
* elke herhaling haalt opnieuw JSON op
* `time.sleep(5)` pauzeert 5 seconden zodat je niet constant requests spamt
* Je kan dit script stoppen met `CTRL+C`


## Stap 4 – Visualisatie op een wereldkaart

In stap 3 kon je longitude en latitude uit JSON halen en op het scherm tonen. In deze stap zetten we die coördinaten om naar een **punt op een wereldkaart**.

### 1. Wereldkaart als achtergrond met `matplotlib.image`

Matplotlib kan een afbeelding inlezen en tonen. Als die afbeelding een wereldkaart is, kunnen we er coördinaten op plotten. We gebruiken een equirectangular wereldkaart (“platte” wereldkaart) uit NASA’s Scientific Visualization Studio. 

Je kan die afbeelding in principe zelf downloaden en in de correcte map zetten. Maar als demonstratie zie je hier hoe het via Python kan. Dit script moet je maar één keer uitvoeren en het zet de afbeelding automatisch in de map waarin je je scripts uitvoert.

```python
import requests

# de URL van de afbeelding die we willen downloaden
img_url = "https://svs.gsfc.nasa.gov/vis/a000000/a002900/a002915/bluemarble-2048.png"
# het pad waar we de afbeelding willen opslaan
img_path = "world.png"

# download de afbeelding
r = requests.get(img_url)
# controleer of de download succesvol was
r.raise_for_status()

# schrijf de afbeelding naar een bestand
with open(img_path, "wb") as f:
    f.write(r.content)

print("Gedownload:", img_path)
```

### 2. Afbeelding tonen met juiste coördinaten

```python
import matplotlib.pyplot as plt
# extra matplotlib import voor afbeeldingen
import matplotlib.image as mpimg

# laad de afbeelding
img = mpimg.imread("world.png")

plt.figure(figsize=(10, 5))
plt.imshow(img, extent=[-180, 180, -90, 90])  # longitude/latitude bereik
plt.xlim(-180, 180)
plt.ylim(-90, 90)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Wereldkaart als achtergrond")
plt.show()
```

Nieuwe onderdelen:

* `matplotlib.image.imread()` leest de afbeelding in.
* `plt.imshow(..., extent=[...])` legt vast welke data-coördinaten bij de randen van de afbeelding horen.

Voor een equirectangular wereldkaart nemen we:

* x-as = longitude van **-180** tot **180**
* y-as = latitude van **-90** tot **90**

### 3. ISS-punt plotten op de kaart

Herinner: bij de ISS-API zijn latitude/longitude **strings**, dus voor plotten zetten we om naar `float`.

```python
import requests
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 1) data ophalen
url = "http://api.open-notify.org/iss-now.json"
data = requests.get(url).json()

lat_str = data["iss_position"]["latitude"]
lon_str = data["iss_position"]["longitude"]

lat = float(lat_str)
lon = float(lon_str)

# 2) kaart tonen
img = mpimg.imread("world.png")
plt.figure(figsize=(10, 5))
plt.imshow(img, extent=[-180, 180, -90, 90])
plt.xlim(-180, 180)
plt.ylim(-90, 90)

# 3) punt toevoegen
plt.scatter(lon, lat, color="red", s=100, label="ISS")
plt.title("ISS positie (1 meting)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.show()
```
### 4. ISS-positie elke 5 seconden updaten

Om de kaart niet telkens opnieuw te openen, maken we:

* één figuur
* één achtergrond
* één punt (scatter-object)
* en werken we enkel de positie bij

Daarvoor gebruiken we **interactieve modus** van matplotlib.

#### Nieuwe concepten

* `plt.ion()` → interactieve modus aanzetten
* `set_offsets()` → positie van het punt aanpassen
* `plt.pause()` → korte pauze zodat de figuur kan verversen

#### Volledig script (live update)

```python
import requests
import time
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Interactieve modus inschakelen
plt.ion()

# 1) Figuur en kaart één keer maken
img = mpimg.imread("world.png")

fig, ax = plt.subplots(figsize=(10, 5))
ax.imshow(img, extent=[-180, 180, -90, 90])
ax.set_xlim(-180, 180)
ax.set_ylim(-90, 90)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title("Live ISS positie")

# 2) Eerste punt aanmaken (start op 0,0)
point = ax.scatter(0, 0, color="red", s=100)

url = "http://api.open-notify.org/iss-now.json"

while True:
    # Nieuwe data ophalen
    data = requests.get(url).json()

    lat = float(data["iss_position"]["latitude"])
    lon = float(data["iss_position"]["longitude"])

    # Puntpositie aanpassen
    point.set_offsets([lon, lat])

    # Figuur verversen
    plt.pause(0.1)

    # 5 seconden wachten
    time.sleep(5)
```

#### Wat gebeurt hier precies?

**We tekenen de kaart maar één keer**

```python
fig, ax = plt.subplots()
```

Hier maken we een `Axes`-object (`ax`).
Vanaf nu werken we altijd via `ax`.

**We maken één scatter-object**

```python
point = ax.scatter(0, 0, ...)
```

Dat object bewaren we in de variabele `point`.

---

**We wijzigen enkel de coördinaten**

```python
point.set_offsets([lon, lat])
```

Hier wordt het bestaande punt verplaatst. Er wordt dus geen nieuw punt getekend.


**Waarom `plt.pause()` nodig is**

Matplotlib moet even de tijd krijgen om de figuur te hertekenen. Zonder `plt.pause()` zie je geen update.

## Uitbreiding: correctere kaarten met Cartopy

De methode met een achtergrondafbeelding is handig, maar heeft beperkingen:

* De kaart is “gewoon een foto” met een vaste rechthoekige projectie.
* Je hebt geen echte kaartfeatures (kustlijnen, grenzen) als “data”.
* Projecties (en het correct omgaan met de randen van de wereld) kan problematisch worden.

**Cartopy** werkt met echte kaartprojecties en geodata op een correcte manier (en integreert met matplotlib).

### Installatie

```bash
pip install cartopy
```

### Voorbeeld: ISS plotten met Cartopy

```python
import requests
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

# ISS data ophalen
url = "http://api.open-notify.org/iss-now.json"
data = requests.get(url).json()

lat = float(data["iss_position"]["latitude"])
lon = float(data["iss_position"]["longitude"])

# Kaart maken (PlateCarree = lat/lon “platte” projectie)
fig = plt.figure(figsize=(10, 5))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.stock_img()       # simpele achtergrond
ax.coastlines()      # kustlijnen

# Punt toevoegen (let op transform)
ax.plot(lon, lat, marker="o", transform=ccrs.PlateCarree())

ax.set_title("ISS positie (Cartopy)")
plt.show()
```

Waarom is dit “beter”?

* Je tekent niet op een foto, maar op een **GeoAxes** met een **projectie**.
* Cartopy weet hoe coördinaten naar de kaart getransformeerd moeten worden.
* Je kan later makkelijk wisselen van projectie zonder je data te veranderen.
